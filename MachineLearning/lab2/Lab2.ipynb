```python
def outlines(df, p):
    Q1=df["points"].quantile(0.25)
    Q3=df["points"].quantile(0.75)
    IQR=Q3-Q1
    lowqe_bound=Q1 - (1.5 * IQR)
    upper_bound=Q3 + (1.5 * IQR)
    ls=df.index[(df[p]>upper_bound)|(df[p]<lowqe_bound)]
    return ls;
```

```python
index_list=[];
for point in ['points']:
    index_list.extend(outlines(df, point))
```

```python
def remove(df, ls):
    ls=sorted(set(ls))
    df=df.drop(ls)
    return df
```

```python
df_r=remove(df,index_list)
```

```python
df_r["points"].plot(kind="box")
```

```python
df_encoded = df_r.apply(lambda x: pd.factorize(x)[0])
```
